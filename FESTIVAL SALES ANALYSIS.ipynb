```python
ax = sns.countplot(x = 'Gender', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sales_gen
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
bp = sns.barplot(x = 'Gender', y='Amount', data = sales_gen)
for bars in bp.containers:
    bp.bar_label(bars)
```

```python
df.columns
```

```python
ax = sns.countplot(data =df, x= 'Age Group', hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index=False )['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Age Group', y='Amount', data = sales_age)
```

```python
print('From above graphs it is observed that most of the buyers are Females within the age group of 26-35 yrs female')
```

```python
# total amount/sales from all the states

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False)
sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(data=sales_state, x='State', y='Orders')
```

```python
print('From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively')
```

```python
ax = sns.countplot(data = df, x= 'Marital_Status')
sns.set(rc={'figure.figsize':(4,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
print('From above graphs we can see that most of the buyers are married (women) and they have high purchasing power')
```

```python
ax = sns.countplot(data = df, x='Occupation')
sns.set(rc={'figure.figsize':(30,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
print('*From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector*')
```

```python
ax = sns.countplot(data = df, x='Product_Category')
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_prod = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data= sales_prod, x='Product_Category', y='Amount')
```

```python
print('From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category')
```

```python
sales_id = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data= sales_id, x= 'Product_ID', y= 'Orders')
```

```python
print('CONCLUSION : Married women age group 26-35 yrs from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category')
```
